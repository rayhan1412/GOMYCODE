```python
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
import pandas as pd
newdata=pd.read_csv('data.csv')
newdata




```

```python
newdata.drop("Unnamed: 0",axis=1,inplace=True)
newdata
x=newdata.drop(["Survived", "Name",  "Ticket", "Embarked","Title"], axis=1)
y= newdata["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
newdata["Survived"].value_counts()
```

```python
dtree = tree.DecisionTreeClassifier()  
dtree.fit(x_train, y_train)   
y_pred=dtree.predict(x_test)   
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30,20))
plot_tree(dtree,filled=True,feature_names=x.columns,proportion=True)
```

```python

dtree = tree.DecisionTreeClassifier(splitter='best',max_depth=7,max_leaf_nodes=7)  
dtree.fit(x_train, y_train)   
y_pred=dtree.predict(x_test)   
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

this accuracy is higher than the previous one

```python
plt.figure(figsize=(30,20))
plot_tree(dtree,filled=True,feature_names=x.columns,proportion=True)
```

```python
from sklearn import datasets
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn import metrics  
x=newdata.drop(["Survived", "Name",  "Ticket", "Embarked","Title"], axis=1)
y= newdata["Survived"]
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3) 
```

```python
clf=RandomForestClassifier(n_estimators=10)  
clf.fit(x_train, y_train)  
y_pred=clf.predict(x_test)  
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
clf=RandomForestClassifier(n_estimators=50)  
clf.fit(x_train, y_train)  
y_pred=clf.predict(x_test)  
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
clf=RandomForestClassifier(n_estimators=100)  
clf.fit(x_train, y_train)  
y_pred=clf.predict(x_test)  
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

the accuracies of these forest trees with different numbers of estimators are slightly different
